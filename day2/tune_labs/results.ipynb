```python
import matplotlib.pyplot as plt 
import numpy as np


aos = np.loadtxt("aos.gcc.dat")
soa = np.loadtxt("soa.gcc.dat")
```

Format: 
memory footprint in kB
Mflop/s of distance(...)
Mflop/s of distcheck(...)
Mflop/s of total program
runtime in secs

```python
plt.style.use("seaborn")
```

```python
print(aos.shape[0])
```

```python
aos.shape[1]
```

```python
fig, ax = plt.subplots(aos.shape[1]-1,1,sharex="col",)

ylabs = ["Mflop/s of distance","Mflop/s of distcheck","Mflop/s total","runtime/s"]
font = {'color':  'darkred',
        'weight': 'normal',
        'size': 12,
        }


for i in range(1,aos.shape[1]):
    ax[i-1].plot(aos[:,0],aos[:,i],label="AOS")
    ax[i-1].plot(soa[:,0],soa[:,i],label="SOA")
    ax[i-1].set_ylabel(ylabs[i-1],fontdict=font)

ax[-1].set_xlabel("memory footprint/kB")
fig.set_figheight(7)
fig.set_tight_layout(tight=True)
handles, labels = ax[-1].get_legend_handles_labels()
fig.legend(handles, labels, loc='upper right',prop={"size":16})
    
```

SOA is way better :)
